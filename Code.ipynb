# MAIN MENU

```python
import pandas as pd
import matplotlib.pyplot as plt
from question import (loading_through_csv, loading_through_pandas, 
                      retrieve_by_oem_id, retrieve_device_info_by_codename, 
                      retrieve_device_info_by_ram_capacity, retrieve_with_condition, 
                      top_regions_for_brand, average_price_for_brand, 
                      derive_screen_insights, create_ram_type_chart, 
                      create_usb_connector_chart, create_price_trends_charts, 
                      create_scatter_plot,custom_analysis)

def print_table(data):
    if not data:
        print("No data to display.")
        return

    keys = data[0].keys()
    max_lengths = {key: max(len(key), max(len(str(row[key])) for row in data)) for key in keys}

    print(" | ".join(key.ljust(max_lengths[key]) for key in keys))
    print("-+-".join('-' * max_lengths[key] for key in keys))

    for row in data:
        print(" | ".join(str(row[key]).ljust(max_lengths[key]) for key in keys))
    print()

def retrieve_data_task_a(data):
    while True:
        print("\nTask a) Retrieve Data:")
        print("a1. Retrieve data based on oem_id")
        print("a2. Retrieve data based on code name")
        print("a3. Retrieve data based on RAM capacity")
        print("a4. Retrieve data based on a specific condition")
        print("0. Return to main menu")
        choice = input("Enter your choice: ")

        try:
            if choice == 'a1':
                oem_id = input("Enter the OEM ID: ")
                results = retrieve_by_oem_id(data, oem_id)
                print_table(results)
            elif choice == 'a2':
                codename = input("Enter the code name: ")
                results = retrieve_device_info_by_codename(data, codename)
                print_table(results)
            elif choice == 'a3':
                ram_capacity = input("Enter the RAM capacity: ")
                results = retrieve_device_info_by_ram_capacity(data, ram_capacity)
                print_table(results)
            elif choice == 'a4':
                results = retrieve_with_condition(data)
                print_table(results)
            elif choice == '0':
                break
            else:
                print("Invalid choice. Please try again.")
        except Exception as e:
            print(f"An error occurred while retrieving data: {e}")

def analyze_data_task_b(data):
    while True:
        print("\nTask b) Analyze Data:")
        print("b1. Identify the top 5 regions where a specific brand of devices was sold")
        print("b2. Analyse the average price of devices within a specific brand, all in the same currency")
        print("b3. Derive screen insights")
        print("b4. Custom analysis")
        print("0. Return to main menu")
        choice = input("Enter your choice: ")

        try:
            if choice == 'b1':
                brand = input("Enter the brand name: ")
                results = top_regions_for_brand(data, brand)
                print(results)
            elif choice == 'b2':
                brand = input("Enter the brand name: ")
                results = average_price_for_brand(data, brand)
                print(results)
            elif choice == 'b3':
                results = derive_screen_insights(data)
                print(results)
            elif choice == 'b4':
                custom_analysis(data)
            elif choice == '0':
                break
            else:
                print("Invalid choice. Please try again.")
        except Exception as e:
            print(f"An error occurred while analyzing data: {e}")

def visualize_data_task_c(data):
    while True:
        print("\nTask c) Visualize Data:")
        print("c1. Create a chart to represent the proportion of RAM types")
        print("c2. Create a chart to compare the number of devices for each USB connector type")
        print("c3. Create charts illustrating the monthly average price trends for 2020-2023")
        print("c4. Create a custom visualization (Scatter Plot: Device Weight vs. Price)")
        print("0. Return to main menu")
        choice = input("Enter your choice: ")

        try:
            if choice == 'c1':
                create_ram_type_chart(data)
            elif choice == 'c2':
                create_usb_connector_chart(data)
            elif choice == 'c3':
                create_price_trends_charts(data)
            elif choice == 'c4':
                create_scatter_plot(data)
            elif choice == '0':
                break
            else:
                print("Invalid choice. Please try again.")
        except Exception as e:
            print(f"An error occurred while visualizing data: {e}")

def main_menu():
    data = []
    file_location = "device_features.csv"  # Relative path to the CSV file
    pandas_data = pd.DataFrame()

    while True:
        print("\nMain Menu:")
        print("a. Load and retrieve data using CSV module")
        print("b. Analyze data using Pandas module")
        print("c. Visualize data using Matplotlib module")
        print("0. Exit")
        choice = input("Enter your choice: ")

        try:
            if choice == 'a':
                data = loading_through_csv(file_location)
                if data:
                    retrieve_data_task_a(data)
            elif choice == 'b':
                if pandas_data.empty:
                    pandas_data = loading_through_pandas(file_location)
                if not pandas_data.empty:
                    analyze_data_task_b(pandas_data)
            elif choice == 'c':
                if pandas_data.empty:
                    pandas_data = loading_through_pandas(file_location)
                if not pandas_data.empty:
                    visualize_data_task_c(pandas_data)
            elif choice == '0':
                break
            else:
                print("Invalid choice. Please try again.")
        except FileNotFoundError:
            print("The specified file was not found. Please check the file path and try again.")
        except pd.errors.EmptyDataError:
            print("No data found in the file. Please check the file content and try again.")
        except pd.errors.ParserError:
            print("Error parsing the file. Please check the file format and try again.")
        except Exception as e:
            print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    main_menu()
```
